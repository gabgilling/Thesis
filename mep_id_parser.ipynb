```python
for i in range(len(meps)):

    url = 'https://parltrack.org/mep/' + unidecode.unidecode(re.sub(' ', '', meps['Name'].iloc[i]))
    try:
        html_page = urlopen(url)
        soup = BeautifulSoup(html_page)
        img_tags = [img['src'] for img in soup.find_all('img')]
        id_in_tag = [line for line in img_tags if '.jpg' in line]
#         print(id_in_tag)
        meps['ID'].iloc[i] = int(''.join(re.findall('\d+', ' '.join(id_in_tag))))
    except:
        print(meps['Name'].iloc[i])
```
